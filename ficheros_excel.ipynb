```python
#Cargar un dataframe a partir de un fichero de excel, para leer un fichero de excel se usa la librería, llamada openpyxl, 
# un dataframe solo cargará una hoja de excel (tendremos que especificar)
%pip install openpyxl
```

```python
import pandas as pd
#debes conectar con cada hoja por separado
agosto2014= pd.read_excel(r"C:\Users\Laura Fort\Documents\Python\introductionPython\VentasGroucery_Meses.xlsx",
                          sheet_name="Ventas Agosto 2014")
agosto2014

septiembre2014= pd.read_excel(r"C:\Users\Laura Fort\Documents\Python\introductionPython\VentasGroucery_Meses.xlsx",
                          sheet_name="Ventas Septiembre 2014")
septiembre2014
```

```python
#Cómo anexar 2 DataFrames para obtener una solo, con todos los registros
#CUIDADO!! deben tener misma estructura

#Importante! siempre ignore_index ya que sino habría IDs repetidas
biMensual= pd.concat([agosto2014, septiembre2014], ignore_index=True)
biMensual

biMensual["Amount"]=biMensual["Unit Price"]*biMensual["Quantity"]
biMensual
```
