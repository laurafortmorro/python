FILE DEDICADO A LOS CONCEPTOS DEL DATAFRAME

```python
#El dataframe de pandas sirve para trabajar con estructuras de 2 dimensiones y se puede encargar de 2 manreas,
#  manialmente o leyendo de un fichero, bade de datos o cualquier origen de datos

import pandas as pd #pd es el acrónimo de pandas
#Dataframe creado manualmente
clase = pd.DataFrame([{"Nombre":"Pedro", "Estatura":1.71,"Edad":28},
                      {"Nombre":"María", "Estatura":1.73,"Edad":45},
                      {"Nombre":"Andrea", "Estatura":1.68,"Edad":61},
                      {"Nombre":"Marcos", "Estatura":1.65,"Edad":19},
                      ])

clase


```

```python
#Cómo ver solamente un campo o columna
clase["Edad"]
clase["Nombre"]
clase["Estatura"].sort_values(ascending=True)
```

```python
#Funciones de resumen
clase["Edad"].sum()
print(clase["Estatura"].max())
print(clase["Estatura"].min())

#Promedio de edad, con redondeo
print(round(clase["Edad"].mean(),2))
```

Como añadir columnas a un DataFrame, tenemos 3 opciones:
- Opción A: Poniendo un valor manual fijo
- Opción B: Haciendo un cálculo a partir de una o varias columnas existentes
- Opción C: Aplicando una función con Apply

```python
#OPCIÓN A
#Si el nombre del campo que pones no existe se crea uno con ese nombre de forma automática
#Esta fórmula sirve para valores fijos en todas las entradas
clase["Colegio"]= "IT Academy"
clase
```

```python
#OPCIÓN B
clase["Edad2"]= clase["Edad"]+2
clase
```

```python
clase["Diferencia"]= clase["Edad2"]-clase["Edad"]
clase
```

```python
#Incrementar la estatura de cada alumno en un 10%
clase["Estatura2"]= clase["Estatura"]*1.1   #con este cálculo se incrementa la estarura un 10%
clase
```

```python
#OPCIÓN C: Con apply, restar 2 años a los alumnos que tengan >30 años

def resta2(edad):
    if edad>30:
        return edad-2
    else:
        return edad #se debe crear siempre un else porque sino ese campo quedaría en blanco
    
#Crear la columna nueva con apply
clase["Edad5"]= clase["Edad"].apply(resta2)
clase
```

```python
#Cómo eliminar una columna, con la instrucción DROP + el parámetro AXIS
clase= clase.drop(["Estatura2"], axis=1) #Axis 1 significa eliminar columnas y Axis 2 significa eliminar filas
clase
```

```python
clase.drop(["Edad2"], axis= 1, inplace=True) #sirve para reescribir la DataFrame con las nuevas indicaciones
clase
```

```python
#Para eliminar una fila, se usa igualmente DROP, pero esta vez con axis= 0
clase.drop([2], axis= 0, inplace=True)
clase
```
